```python
import matplotlib.pyplot as plt
import os
import glob
import cv2
import numpy as np
```

```python
target_width, target_height = 200, 150
cv2.destroyAllWindows()
```

```python
def read_images(path):
    image_files = glob.glob(os.path.join(path, "*.jpeg"))
    images = []

    for image_file in image_files:
        image = cv2.imread(image_file)
        image = cv2.resize(image, (target_width, target_height))
        images.append(image)
    return images
```

```python
test_normal = read_images("chest_xray/chest_xray/test/normal")
test_pneumonia = read_images("chest_xray/chest_xray/test/pneumonia")
test_X = test_normal + test_pneumonia
test_y = np.zeros(len(test_normal)).tolist() + np.ones(len(test_pneumonia)).tolist()
```

```python
fig, ax = plt.subplots()

ax.bar(['normal', 'pneumonia'], [len(test_normal), len(test_pneumonia)])

ax.set_ylabel('Number of patients')

ax.set_title('Patients')

plt.show()
```

```python
num_rows = 5
num_cols = 5
fig = plt.figure(figsize=(10, 10))

for i in range(num_rows * num_cols):
    if i >= 25:
        break
    img = test_normal[i]
    subplt = fig.add_subplot(num_rows, num_cols, i + 1)
    subplt.axis("off")
    plt.imshow(img)
plt.show()
```
