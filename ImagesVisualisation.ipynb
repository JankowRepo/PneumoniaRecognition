```python
from keras import layers
from keras.optimizers import SGD, Adam
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
import datetime
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import winsound
from sklearn.utils import resample
import time
from IPython.display import clear_output
import seaborn as sb
import matplotlib.pyplot as plt
import os
import glob
import cv2
```

```python
normal_images_directory = "ChestXRay2017/chest_xray/test/normal"
normal_image_files = glob.glob(os.path.join(normal_images_directory, "*.jpeg"))
normal_images = []
target_width, target_height = 2000, 1500

for image_file in normal_image_files:
    normal_img = cv2.imread(image_file)
    normal_img = cv2.resize(normal_img, (target_width, target_height))
    normal_images.append(normal_img)
```

```python
pneumonia_images_directory = "ChestXRay2017/chest_xray/test/pneumonia"
pneumonia_image_files = glob.glob(os.path.join(pneumonia_images_directory, "*.jpeg"))
pneumonia_images = []
target_width, target_height = 2000, 1500

for image_file in pneumonia_image_files:
    pneumonia_img = cv2.imread(image_file)
    pneumonia_img = cv2.resize(pneumonia_img, (target_width, target_height))
    pneumonia_images.append(pneumonia_img)
```

```python
fig, ax = plt.subplots()

ax.bar(['normal', 'pneumonia'], [len(normal_images), len(pneumonia_images)])

ax.set_ylabel('Number of patients')

ax.set_title('Patients')

plt.show()
```

```python
num_rows = 5
num_cols = 5
fig = plt.figure(figsize=(10, 10))

for i in range(num_rows * num_cols):
    if i >= 25:
        break
    img = normal_images[i]
    subplt = fig.add_subplot(num_rows, num_cols, i + 1)
    subplt.axis("off")
    plt.imshow(img)
plt.show()
```

```python
num_rows = 5
num_cols = 5
fig = plt.figure(figsize=(10, 10))

for i in range(num_rows * num_cols):
    if i >= 25:
        break
    img = pneumonia_images[i]
    subplt = fig.add_subplot(num_rows, num_cols, i + 1)
    subplt.axis("off")
    plt.imshow(img)
plt.show()
```
